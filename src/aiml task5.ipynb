```python
#Task 1 – Load CSV, display first 5 rows, check missing values, summary statistics:
import pandas as pd
import numpy as np

df = pd.read_csv('top_books.csv')
print(df.head())
print('\nMissing Values:\n', df.isnull().sum())
print('\nSummary Statistics:\n', df.describe())

#Task 2 – Fill missing values with column median:
for col in df.select_dtypes(include=np.number).columns:
    df[col].fillna(df[col].median(), inplace=True)
print('Missing values after fill:\n', df.isnull().sum())

#Task 3 – Filter rows where a numerical column > threshold:
col = df.select_dtypes(include=np.number).columns[0]  # First numeric col
threshold = df[col].mean()                             # Using mean as threshold
filtered = df[df[col] > threshold]
print(filtered.head())

#Task 4 – Group by categorical column, compute mean and std:
cat_col = df.select_dtypes(include='object').columns[0]
grouped = df.groupby(cat_col).agg(['mean', 'std'])
print(grouped)

#Task 5 – Merge two DataFrames on a common column:
df1 = df.iloc[:, :3].copy()
df2 = df.iloc[:, [0, 3, 4]].copy()
common_col = df.columns[0]
merged = pd.merge(df1, df2, on=common_col)
print(merged.head())

#Task 6 – Create new column from existing columns:
num_cols = df.select_dtypes(include=np.number).columns
df['new_col'] = df[num_cols[0]] + df[num_cols[1]]
print(df.head())

#Task 7 – Correlation matrix:
corr_matrix = df.select_dtypes(include=np.number).corr()
print(corr_matrix)

#Task 8 – Scatter plot:
import matplotlib.pyplot as plt

x_col, y_col = num_cols[0], num_cols[1]
plt.scatter(df[x_col], df[y_col], alpha=0.6, color='steelblue')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title(f'Scatter Plot: {x_col} vs {y_col}')
plt.tight_layout()
plt.show()
```
